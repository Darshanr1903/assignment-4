```python
import numpy as np
import pandas as pd
import plotly.express as px

# Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Aggregate sales by Month and Day
sales_matrix = df.groupby(['Month', 'Day'])['Sales'].sum().unstack(fill_value=0)

# Create the heatmap
fig = px.imshow(
    sales_matrix.values,
    x=sales_matrix.columns,
    y=sales_matrix.index,
    labels={'x': 'Day of Month', 'y': 'Month', 'color': 'Total Sales'},
    title='Sales Heatmap by Month and Day'
)

# Show the heatmap
fig.show()
```
